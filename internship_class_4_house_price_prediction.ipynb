```python
# step 1 : import libraries
import pandas as pd
```

```python
# step 2 : import data or read data
house = pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/Boston.csv')
```

```python
house.head()
```

```python
# step 3 : define your target (y) dependent varible and your features (X) independent variable
house.columns
```

```python
# syntax to define your model
y = house['MEDV']
X =house[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT']]
# where salary is dependent upon experience
```

```python
# step 4 :  Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state = 2529)
X_train
```

```python
# step 5 : select model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# step 6 : train model (fit model)
model.fit(X_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
# step 7: prediction
y_pred = model.predict(X_test)
```

```python
# step 8 : accuracy
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(y_test,y_pred)
```
