```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size = 0.7, random_state = 2529)
X_train
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

Now the sample is random but it is fixed  now we have completed the train-test-splitmodel and we have also completed a new library i.e. sklearn

```python
# step 5 : select model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# step 6 : train model (fit model)
model.fit(X_train,y_train)
```

```python

model.intercept_
```

```python



model.coef_
```

formula to predict salary :  Salary = 26596.9 + 9405.6 Experience

```python
# step 7: prediction
y_pred = model.predict(X_test)
```
